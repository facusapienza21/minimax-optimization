```julia
include("minimax.jl")
include("auxiliar.jl")

using Plots
using LaTeXStrings
```

```julia
L1(x::Vector) = x[1] * x[2]; 
L2(x::Vector) = x[1]^2 - 10000 * x[2]^2;
L3(x::Vector) = (x[1] ) * (x[2]) + (1 / 3) * exp( - (x[1])^2 - (x[2])^2 );
L4(x::Vector) = ( x[1]^4 * x[2]^2 + x[1]^2 + 1 ) * ( x[1]^2 * x[2]^4 - x[2]^2 + 1);
L5(x::Vector) = (x[1] - 0.5) * (x[2] - 0.5) + (1 / 3) * exp( - (x[1] - 0.5)^2 - (x[2] - 0.75)^2 );
```

## Tolerance as a function of alpha

```julia
function tolerance(z, eps, z0)
   
    for i in 1:length(z)
        if norm(z[i] - z0) < eps
            return i
        end
    end
    
    return length(z)
    
    end;
```

```julia
f = L5
z0 = [0, 0]
init = [1, 1]

try_alpha = 10 .^ [-2:0.05:2;]

epsilon = 1e-5
N = 130000

T_OMD = []
T_IMP = []
T_IMP2 = []
T_IOD = []
T_FAC = []

for alfa in try_alpha
   
    OMD  = minimax_solver(f, "OMD", 1, 1, N, alfa, init) 
    IMP  = minimax_solver(f, "ITD", 1, 1, N, alfa, init, 1) 
    IMP2 = minimax_solver(f, "ITD", 1, 1, N, alfa, init, 2)
    #IOD  = minimax_solver(f, "IOD", 1, 1, N, alfa, init, 1)
    FAC  = minimax_solver(f, "227", 1, 1, N, alfa, init, 1) 
    
    push!(T_OMD, tolerance(OMD.z, epsilon, z0) )
    push!(T_IMP, tolerance(IMP.z, epsilon, z0) )
    push!(T_IMP2, tolerance(IMP2.z, epsilon, z0) )
    #push!(T_IOD, tolerance(IOD.z, epsilon, z0))
    push!(T_FAC, tolerance(FAC.z, epsilon, z0))
    
end
```

```julia
gr(xlims = (-2.1,2.1), ylim = (0, 5.5), legend=:topright, xlabel = L"\log_{10} \alpha", ylabel = L"\log_{10}T(\epsilon, \alpha)")
plot(log10.(try_alpha), log10.(T_OMD), markershape = :circle, color = mycolors[1], ms = 2, lw = 1.5, label = "OMD" )
plot!(log10.(try_alpha), log10.(T_IMP), markershape = :circle, color = mycolors[2], ms = 2, lw = 1.5, label = "ITD" )
plot!(log10.(try_alpha), log10.(T_IMP2), markershape = :circle, color = mycolors[3], ms = 2, lw = 1.5, label = "2-ITD" )
#plot!(log10.(try_alpha), log10.(T_IOD), markershape = :circle, color = mycolors[4], ms = 2, lw = 1, label = "IOD" )
plot!(log10.(try_alpha), log10.(T_FAC), markershape = :circle, color = mycolors[4], ms = 2, lw = 1.5, label = "227" )
```
